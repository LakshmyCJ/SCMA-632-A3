# PROBIT REGRESSION ANALYSIS

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.discrete.discrete_model import Probit
```

```python
df=pd.read_csv('/content/NSSO68.csv')
```

```python
# Create the dependent variable
df['non_vegetarian'] = np.where(df['nonvegtotal_q'] > 0, 1, 0)
```

```python
features = ['hhdsz', 'Religion', 'Social_Group', 'Type_of_land_owned', 'Land_Owned',
            'MPCE_URP', 'Age', 'Sex', 'Education', 'Regular_salary_earner', 'nonvegtotal_q']

df = df[features]
```

```python
df = df.dropna()

# Define X (features) and y (target)
X = df.drop('nonvegtotal_q', axis=1)
y = np.where(df['nonvegtotal_q'] > 0, 1, 0)
```

```python
# Add constant to the independent variables
X = sm.add_constant(X)

# Fit probit regression model
probit_model = sm.Probit(y, X)
probit_results = probit_model.fit()

# Print summary statistics
print(probit_results.summary())
```
